```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df
```

```python
sns.countplot(x = df['sex'], hue = df['smoker'])
```

```python
sns.barplot(x = df['sex'], y= df['total_bill'])
```

```python
import numpy as np
sns.barplot(x = df['sex'], y= df['tip'], estimator=np.sum)
```

```python
sns.boxplot(x = 'tip', y = 'day', data= df, palette='rainbow')
```

```python
sns.violinplot(x = df['day'], y = df['total_bill'], data= df)
```

```python
sns.stripplot(x = df['day'], y = df['total_bill'], data=df)
```

```python
sns.swarmplot(x = df['day'], y = df['total_bill'], data=df)
```

```python
sns.violinplot(x = df['tip'], y = df['day'], data=df)
```

```python
sns.violinplot(x = df['tip'], y = df['day'], data=df)
sns.swarmplot(x = df['tip'], y = df['day'], data=df)
```
