```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Matplotlib basics

x = np.linspace(0,5, 11)
x
y = x**3
y
```

```python
plt.plot(x, y)
plt.title("Hello")
plt.xlabel("X axis")
plt.ylabel("y axis")
```

```python
# Subplots in matplotlIb
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(x, y)
plt.subplot(2,2,3)
plt.plot(y, x)
plt.subplot(2,2,2)
plt.plot(x**2, x**2)

```

```python
#Object Oriented way

fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,1,1])
axes1.plot(x,y)
axes2 = fig.add_axes([0.4,0.5,0.5,0.4])
axes2.plot(x,y)
```

```python
#TYpes of plots
plt.scatter(x, y)
```

```python
from random import sample
data = sample(range(1,1000),100)
plt.hist(data)
```

```python
plt.boxplot(x)
```

```python
#saving plots

fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("BasicPLot.png")
```
